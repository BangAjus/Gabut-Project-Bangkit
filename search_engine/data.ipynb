# Import all the requirements

```python
from bs4 import BeautifulSoup as bs
import requests
import re
import pandas as pd
```

# Make the user-agent

```python
header = {
    'user-agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36"
}
```

# Make function of browse and find the data

```python
def browse_and_find(keyword):

    keyword = keyword.lower().replace(" ", "-")
    # browse with keyword 
    url = f"https://www.jobstreet.co.id/id/{keyword}-jobs"
    req = requests.get(url, headers = header)
    sup = bs(req.text, 'html.parser')

    title =  [i.text for i in sup.find_all('div', 'z1s6m00 l3gun70 l3gun74 l3gun72')]
    region = [i.text for i in sup.find_all('span', 'z1s6m00 _1hbhsw64y y44q7i0 y44q7i3 y44q7i21 y44q7ih') if re.search(r'[0-9]+', i.text) == None]
    company = [i.text for i in sup.find_all('span', 'z1s6m00 bev08l1 _1hbhsw64y _1hbhsw60 _1hbhsw6r') if i.text not in region][5:35]
    category = [keyword]*len(title)

    data = {'title':title,
            'region':region,
            'company':company,
            'category':category}
    
    return data
```

# you can process the data manually without function like this

```python
data1 = browse_and_find('software engineer')
data2 = browse_and_find('data science')
data3 = browse_and_find('ui ux design')
```

```python


data1 = pd.DataFrame(data1)
data2 = pd.DataFrame(data2)
data3 = pd.DataFrame(data3)

data = pd.concat([data1, data2, data3], ignore_index=True)
data
```

# Or, you can make a function like this to save more energy

```python
def frame_all_data(keywords):

    data_list = []
    for i in keywords:
        data = browse_and_find(i)
        data_list.append(pd.DataFrame(data))

    return pd.concat([i for i in data_list], ignore_index=True)
```

```python
data = frame_all_data(['software engineer', 'data science', 'ui ux design'])
data
```

# The results are the same, now make the csv file

```python
data.to_csv('data.csv', index=False)
```
